## Generate the plot

### Subtask:
Create the code to generate the chosen plot using a suitable plotting library like Matplotlib or Seaborn.


**Reasoning**:
Create a scatter plot of 'bmi' vs. 'charges' using matplotlib, add a title and labels, and display the plot.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(insurance_df['bmi'], insurance_df['charges'])
plt.title('Scatter Plot of BMI vs. Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()
```
